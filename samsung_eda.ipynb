##1.Dataset Loading

This section loads the dataset using pandas.

```python
import pandas as pd
df=pd.read_csv("../data/Expanded_Dataset.csv")
df
```

##2.Data Overview

In this section, we examine the basic structure of our dataset, including:

-**Shape**: Number of rows and columns

-**Columns**: List of all column names

-**Data Types & Missing Values**: Check for nulls and data types

-**Preview of Data**: First and last few rows

-**Summary Statistics**: Numeric summary of the dataset

```python
print("Shape of the dataset:", df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nDataset Info:")
df.info()

print("\nFirst 5 rows:")
display(df.head())

print("\nLast 5 rows:")
display(df.tail())

print("\nSummary Statistics:")
display(df.describe())
```

##3.Data Cleaning

Before analyzing the data, we need to check for and handle any issues such as:

-Missing values

-Duplicate rows

-Incorrect or inconsistent data types

-Any other anomalies in the dataset

```python
print("Missing values in each column:")
print(df.isnull().sum())

print("\nNumber of duplicate rows:",
      df.duplicated().sum())

df=df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)

print("\nData Types of each column:")
print(df.dtypes)

print("\nPreview of Cleaned data:")
display(df.head())

```

##4.Exploratory Data Analysis (EDA)

In this section, we explore the Samsung smartphone dataset to understand:

-Distribution of key numerical variables such as revenue and units sold

-Market share and 5G related metrics

-Comparison across different regions

-Relationships between important variables

-Analyze trends for revenue, units sold, and 5G subscribers

```python
#Revenue Distribution

import matplotlib.pyplot as plt
plt.figure()
plt.hist(df['Revenue ($)'],color='skyblue',edgecolor='black')
plt.xlabel('Revenue ($)')
plt.ylabel('Frequency')
plt.title('Distribution of Revenue')
plt.show()
```

```python
#Units sold by region 

plt.figure()
df.groupby('Region')['Units Sold'].sum().plot(
    kind='bar',
    color='steelblue',
    edgecolor='black'
)
plt.xlabel('Region')
plt.ylabel('Total Units Sold')
plt.title('Units Sold by Region')
plt.show()
```

```python
#Market Share vs Revenue 

plt.figure()
plt.scatter(
    df['Market Share (%)'],
    df['Revenue ($)'],
    color='purple',
    alpha=0.7
)
plt.xlabel('Market Share (%)')
plt.ylabel('Revenue($)')
plt.title('Market Share vs Revenue')
plt.show()
```

```python
#5G Coverage vs Preference for 5G

plt.figure()
plt.scatter(
    df['Regional 5G Coverage (%)'],
    df['Preference for 5G (%)'],
    color='darkorange',
    alpha=0.7
)
plt.xlabel('Regional 5G Coverage(%)')
plt.ylabel('Prefernce for 5G')
plt.title('5G Coverage vs Preference for 5G')
plt.show()
```

```python
#Revenue Contribution by Region

colors=['#4CAF50','#2196F3','#FF9800','#9C27B0','#F44336']
plt.figure()
df.groupby('Region')['Revenue ($)'].sum().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    colors=colors
)
plt.title('Revenue Contribution by Region')
plt.show()
           
```

```python
#Average 5G Speed by Region

plt.figure()
df.groupby('Region')['Avg 5G Speed (Mbps)'].mean().plot(
    kind='bar',
    color='teal',
    edgecolor='black'
)
plt.xlabel('Region')
plt.ylabel('Avg 5G Speed (mbps)')
plt.title('Average 5G Speed by Region')
plt.show()
```

```python
#5G Subscribers Trend

plt.figure()
df.groupby('Year')['5G Subscribers (millions)'].sum().plot(
    color='crimson',
    linewidth=2
)
plt.xlabel('Year')
plt.ylabel('5G Subscribers (millions)')
plt.title('5G Subscribers Trend Over Time')
plt.show()
```

```python
#Revenue Trend Over Time

plt.figure()
df.groupby('Year')['Revenue ($)'].sum().plot(
    color='royalblue',
    linewidth=2
)
plt.xlabel('Year')
plt.ylabel('Revenue ($)')
plt.title('Revenue Trend Over Time')
plt.show()
```

```python
#Correlation Heatmap

import seaborn as sns
numeric_df=df.select_dtypes(include='number')
corr=numeric_df.corr()
plt.figure(figsize=(10,6))
sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    linewidth=0.5
)
plt.title('Correlation Heatmap')
plt.show()
```
